# 3. JFK flights arrive weekday and weekend

```python
# x= 930, 9:30; 2245,22:45
def timezone(x):
    x1 = (x-x%100)/100
    x2 = x%100
    return ((4*x1-1)+ math.floor((x2/15))+1)
```

```python
fullday = [31,28,31,30,31,30]
weekend = [8,8,9,9,8,9]
weekday = (np.array(fullday)-np.array(weekend)).tolist()
print weekday
```

```python
jfk_jan = pd.read_csv('JFK,Jan,Dest,10218.csv')
jfk_feb = pd.read_csv('JFK,Feb,Dest,9397.csv')
jfk_mar = pd.read_csv('JFK,Mar,Dest,10937.csv')
jfk_apr = pd.read_csv('JFK,Apr,Dest,10846.csv')
jfk_may = pd.read_csv('JFK,May,Dest,11240.csv')
jfk_jun = pd.read_csv('JFK,Jun,Dest,11321.csv')
jfk_flight=pd.concat([jfk_jan,jfk_feb,jfk_mar,jfk_apr,jfk_may,jfk_jun],ignore_index=True)
values = {'ArrTime': 10000}
jfk_flight = jfk_flight.fillna(value=values)
jfk_flight_weekday = jfk_flight[(jfk_flight['DayOfWeek']>0) & (jfk_flight['DayOfWeek']<6)]
jfk_flight_weekend = jfk_flight[(jfk_flight['DayOfWeek']>5) & (jfk_flight['DayOfWeek']<8)]
print float(len(jfk_flight_weekday))/len(jfk_flight)*7
print float(len(jfk_flight_weekend))/len(jfk_flight)*7
jfk_flight_weekday_average = [0.0]*96
jfk_flight_weekday_count = [0]*96
jfk_flight_weekend_average = [0.0]*96
jfk_flight_weekend_count = [0]*96
flight_weekday_starttime = jfk_flight_weekday['ArrTime'].tolist()
flight_weekend_starttime = jfk_flight_weekend['ArrTime'].tolist()
# weekday
x =  np.arange(96).tolist()
for i in range(len(flight_weekday_starttime)):
    index = timezone(flight_weekday_starttime[i])
    if index<96:
        jfk_flight_weekday_average[int(index)] += 1
for i in range(96):
    jfk_flight_weekday_average[i] = float(jfk_flight_weekday_average[i])/sum(np.array(weekday))
# weekend
for i in range(len(flight_weekend_starttime)):
    index = timezone(flight_weekend_starttime[i])
    if index<96:
        jfk_flight_weekend_average[int(index)] += 1
for i in range(96):
    jfk_flight_weekend_average[i] = float(jfk_flight_weekend_average[i])/sum(np.array(weekend))
```
